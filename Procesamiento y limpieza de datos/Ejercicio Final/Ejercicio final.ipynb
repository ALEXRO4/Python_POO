```python
import pandas
```

```python
db = pandas.read_csv("https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv")
db
```

```python
db['age'] =  round(db.age.fillna(db.age.mean()),1) 
db
```

```python
db['embarked'] = db.embarked.fillna('S')
db
```

```python
db['sex'] = db['sex'].replace({ 'female': 1, 'male': 0 })
db
```

```python
db['embarked'] = db['embarked'].replace({ 'S': 0, 'C': 1, 'Q': 2 })
db
```

```python
db["child"] = 0
db.loc[db["age"] < 16, "child"] = 1
db
```

```python
df = db.drop(columns=['name','ticket', 'cabin'])
df
```

```python
db_dup=db.duplicated(subset=['name','survived','age','sex','embarked'],keep="last",)
db['dup'] = db_dup
db
```
